# **Encoding**

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
```

```python
label_encoder = LabelEncoder()
df['CourseCategory'] = label_encoder.fit_transform(df['CourseCategory'])
```

```python
df.head()
```

# **Checking Imbalance Data on Target**

```python
df['CourseCompletion'].value_counts()
```

# **Setup Preprocessing with PyCaret**

```python
clf = setup(df, target = 'CourseCompletion',
            preprocess = True,
            fix_imbalance=True,
            normalize=True,
            normalize_method= 'minmax',
            remove_outliers=True,
            train_size =0.6,
            #feature_selection = True,
            use_gpu=True)
```

```python
get_config('X_train_transformed')
```

# **Compares Model**

```python
from pycaret.classification import compare_models
```

```python
best_model = compare_models()
```

**Choosing Model**

```python
rf = create_model('rf')
```

# **Tuning Hyperparameters**

```python
#tune hyperparameters of rf
tuned_rf = tune_model(rf)
```

```python
print(tuned_rf)
```

# **Plotting and Analyze Model Perform**

```python
# Plot confusion matrix
plot_model(rf, plot='confusion_matrix')
```

```python
# Plot ROC curve
plot_model(rf, plot='auc')
```

```python
# plot feature importance
plot_model(rf, plot = 'feature')
```

# **Finalize model**

```python
final_rf = finalize_model(rf)
```

```python
final_rf
```

```python
test_predictions = predict_model(rf, data=df)
```
